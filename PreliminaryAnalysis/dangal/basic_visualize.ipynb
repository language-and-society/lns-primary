```python
import json
import math
import matplotlib.pyplot as plt
```

```python
with open("dangal_character_data.json") as f:
    character_data = json.load(f)
```

```python
for key in character_data.keys():
    print(key, character_data[key]['gender'])
```

```python
words = {}
for key in character_data.keys():
    for sentence in character_data[key]['utterances']:
        for word in sentence.split():
            try:
                words[word.lower()] += 1
            except:
                words[word.lower()] = 1
print(f"Number of unique words: {len(words)}")
```

```python
sorted_words = [(key, value) for key, value in sorted(words.items(), key=lambda item: item[1], reverse=True)]
```

```python
labels = []
values = []
log_values = []
for item in sorted_words:
    labels.append(item[0])
    values.append(item[1])
    log_values.append((math.log(item[1])))
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(labels[:50], values[:50])
fig.autofmt_xdate()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(labels[:50], log_values[:50])
fig.autofmt_xdate()
plt.show()
```

```python
male_utterances = {}
female_utterances = {}

for key in character_data.keys():
    for sentence in character_data[key]['utterances']:
        for word in sentence.split():
            if character_data[key]['gender'] == 'M':
                try:
                    male_utterances[word.lower()] += 1
                except:
                    male_utterances[word.lower()] = 1
            else:
                try:
                    female_utterances[word.lower()] += 1
                except:
                    female_utterances[word.lower()] = 1
```

```python
print("Unique Words said by Males: ", len(male_utterances))
print("Unique Words said by Females: ", len(female_utterances))
```

```python
male_sorted_words = [(key, value) for key, value in sorted(male_utterances.items(), key=lambda item: item[1], reverse=True)]
female_sorted_words = [(key, value) for key, value in sorted(female_utterances.items(), key=lambda item: item[1], reverse=True)]
```

```python
labels = []
values = []
log_values = []
for item in male_sorted_words:
    labels.append(item[0])
    values.append(item[1])
    log_values.append((math.log(item[1])))
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(labels[:50], values[:50])
fig.autofmt_xdate()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(labels[:50], log_values[:50])
fig.autofmt_xdate()
plt.show()
```
